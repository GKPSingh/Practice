```python
# Finding the mean and mode value of the age as per occupation
def get_mode(series):
    mode_vals = series.mode()
    return mode_vals.iloc[0] if not mode_vals.empty else None

# Group and calculate stats
age_stats = df_cust.groupby('occupation')['age'].agg(
    mean='mean',
    median='median',
    mode=get_mode
).reset_index()

# Optional: round numeric values
age_stats[['mean', 'median']] = age_stats[['mean', 'median']].round(2)

# Show result
print(age_stats)
```

```python
# Filling the outlier age values with the mode value of the age as per occupation
def get_mode(series):
    mode_vals = series.mode()
    return mode_vals.iloc[0] if not mode_vals.empty else None

# Compute mode age per occupation
mode_age_by_occ = df_cust.groupby('occupation')['age'].transform(get_mode)

# Define outlier condition
outliers = (df_cust['age'] < 18) | (df_cust['age'] > 80)

# Replace outlier ages with occupation-specific mode
df_cust.loc[outliers, 'age'] = mode_age_by_occ[outliers]
```

```python
# Creating the Credit score range column as per the ranges
bins = [300, 450, 500, 550, 600, 650, 700, 750,800]  # Note: right side is *exclusive* in pd.cut if right=False
labels = [f'{start}-{end-1}' for start, end in zip(bins,bins[1:])] 

# Create a new 'credit_score_group' column
df_cs['credit_score_range'] = pd.cut(df_cs['credit_score'], bins=bins, labels=labels, include_lowest =True, right=False)

df_cs.head()
```
