```python
# convert 'day' column to timestamp
df = pd.read_csv("weather_data2.csv", parse_dates = ["day"])
df
```

```python
type(df['day'][0])
```

```python
df.shape
```

```python
# Setting 'Day' column as an index
df.set_index('day', inplace=True)
df
```

```python
# Filling all null values with zeros
df.fillna(0,inplace=True)
df
```

```python
#Filling Different values for NA in different columns
df.fillna({
    'temperature': df.temperature.mean(),
    'windspeed': df.windspeed.mean(),
    'event': 'No Event'
})
```

```python
# Filling the 'NA' values with forward values
df.fillna(method = 'ffill')
```

```python
# Filling the 'NA' values with back values
df.fillna(method = 'bfill')
```

```python
# filling the 'NA' vaules in desired columns
df.fillna(method='ffill', limit=2)
```

```python
df
```

```python
# filling the linear interpolation of values
df.interpolate()
```

```python
# Dropping the 'NA' values
df.dropna()
```

```python
df
```

```python
# Dropping the values where 'NA' is present in all columns
df.dropna(how='all')
```

```python
# Dropping the rows by setting the limit for 'NA' values
df.dropna(thresh=2)
```
